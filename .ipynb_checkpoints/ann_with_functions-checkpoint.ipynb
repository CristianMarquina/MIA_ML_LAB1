```julia
using Flux
using JLD2, FileIO
using Statistics
using Random
using Printf
using CategoricalArrays
print("1")
# 1. Include Dependencies
include("unit2-multilayer-perceptron.jl")  # ANN Basics
include("unit4-metrics.jl")                # Metrics
include("ann_utils.jl")           # <--- YOUR NEW PIPELINE FILE
print("2")
# 2. Define Topologies to Test (Global Configuration)
topologies_to_test = [
    [8], [16], [32], [64],         
    [16, 8], [32, 16], [64, 32], 
    [8,16],[16,32],[32,64],
    [16, 16], [32, 32], [64,64]     
]

println("--- STARTING FULL EXPERIMENTAL PIPELINE ---\n")

# --------------------------------------------------------------------
# 3. EXECUTE PIPELINE FOR EACH APPROACH
# --------------------------------------------------------------------

# >>> Approach 1: MinMax
# Note: 'process_single_approach' does Loading + Training + Saving automatically.
winner_minmax = process_single_approach(
    "data_checkpoints/approach_1_minmax.jld2", 
    "MinMax", 
    topologies_to_test
)

# >>> Approach 2: PCA
winner_pca = process_single_approach(
    "data_checkpoints/approach_2_pca.jld2", 
    "PCA", 
    topologies_to_test
)

# >>> Approach 3: ICA
winner_ica = process_single_approach(
    "data_checkpoints/approach_3_ica.jld2", 
    "ICA", 
    topologies_to_test
)

# --------------------------------------------------------------------
# 4. FINAL COMPARISON TABLE (OPTIONAL BUT RECOMMENDED)
# --------------------------------------------------------------------
println("\n\n****************************************************************")
println("ðŸ”¥ GRAND FINALE: COMPARISON OF WINNERS PER APPROACH")
println("****************************************************************")
println(" Approach | Best Topology  | Sensib | Acc(%) | F1-Score | Time(s)")
println("----------|----------------|--------|--------|----------|---------")

approaches = [("MinMax", winner_minmax), ("PCA", winner_pca), ("ICA", winner_ica)]

for (name, res) in approaches
    (topo, acc, f1, sens, spec, prec, time_sec, _) = res
    @printf(" %-8s | %-14s | %.4f | %-6.2f | %-8.4f | %.3f\n", 
            name, string(topo), sens, acc, f1, time_sec)
end
```

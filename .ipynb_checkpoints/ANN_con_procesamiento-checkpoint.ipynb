```julia
using MLJ
using MLJMultivariateStatsInterface
using JLD2
using Plots # Aseguramos que Plots est√© cargado para el final

# -------------------------------------------------------------------------
# 1. PREPARACI√ìN Y CARGA DEL MODELO PCA
# -------------------------------------------------------------------------
println("üöÄ INICIANDO APPROACH 2: PCA (RELOADED)...")

# Cargar datos originales
data_path = "heart_disease_uci.csv"
data, num_col, cat_col, target_col = load_and_clean_data(data_path)

# Preprocesamiento Z-Score (Vital para PCA)
approach_2 = prepare_data(
    data, num_col, cat_col, target_col,
    norm_method=:zscore 
)

# --- CORRECCI√ìN DE CARGA DEL MODELO ---
# Usamos 'const' para fijar el modelo y evitar errores de redefinici√≥n
# y lo guardamos en 'MLJ_PCA' para no confundirlo con otros nombres.
const MLJ_PCA = MLJ.@load PCA pkg=MultivariateStats verbosity=0
println("   ‚úÖ Clase PCA cargada correctamente: $MLJ_PCA")

# -------------------------------------------------------------------------
# 2. ENTRENAMIENTO DEL PCA
# -------------------------------------------------------------------------
# Unimos Train + Val para ajustar el PCA (Mejor captura de varianza)
x_train = approach_2.x_train
x_val   = approach_2.x_val
x_test  = approach_2.x_test

x_train_val_combined = vcat(x_train, x_val)
println("   Fitting PCA on $(size(x_train_val_combined)) samples...")

# Configurar modelo (95% varianza)
pca_model = MLJ_PCA(variance_ratio=0.95)

# Entrenar (Fit)
pca_machine = machine(pca_model, MLJ.table(x_train_val_combined))
MLJ.fit!(pca_machine, verbosity=0)

# -------------------------------------------------------------------------
# 3. TRANSFORMACI√ìN Y SEPARACI√ìN
# -------------------------------------------------------------------------
println("   Transforming data...")

# Transformamos todo a tablas reducidas y luego a Matrices para Flux
x_train_val_pca_table = MLJ.transform(pca_machine, MLJ.table(x_train_val_combined))
x_test_pca_table      = MLJ.transform(pca_machine, MLJ.table(x_test))

mat_train_val_pca = MLJ.matrix(x_train_val_pca_table)
mat_test_pca      = MLJ.matrix(x_test_pca_table)

# Recuperamos la separaci√≥n original Train / Val
n_train = size(x_train, 1)
x_train_pca_final = mat_train_val_pca[1:n_train, :]
x_val_pca_final   = mat_train_val_pca[n_train+1:end, :]

println("   ‚úÖ PCA Transform Complete.")
println("      Original Cols: $(size(x_train, 2)) -> PCA Cols: $(size(x_train_pca_final, 2))")

```

```julia
# -------------------------------------------------------------------------
# 4. PUENTE ANN (CON DIAGN√ìSTICO DE DIMENSIONES)
# -------------------------------------------------------------------------

println("\nüîç DIAGN√ìSTICO DE DIMENSIONES (PCA):")
println("   X Train PCA: $(size(x_train_pca_final))")
println("   Y Train OHE: $(size(approach_2.y_train_ohe))")

# Si X tiene m√°s filas que Y (ej: 17 vs 577), es que X est√° transpuesta (Features x Samples)
# trainClassANN quiere (Samples x Features), as√≠ que X debe tener 577 filas.

dataset_pca = (
    # ENTRENAMIENTO
    # Nos aseguramos de pasar Float32
    x_train         = Float32.(x_train_pca_final),
    
    # IMPORTANTE: y_train_encoded debe ser (Samples x Classes)
    # Si approach_2.y_train_ohe ya es as√≠, lo pasamos directo.
    y_train_encoded = approach_2.y_train_ohe,
    
    # Backup (no se usa en v3, pero por compatibilidad)
    y_train_flux    = Float32.(approach_2.y_train_ohe), 
    
    # VALIDACI√ìN
    x_val           = Float32.(x_val_pca_final),
    y_val_encoded   = approach_2.y_val_ohe,
    
    # TEST
    x_test          = Float32.(mat_test_pca),
    y_test_encoded  = approach_2.y_test_ohe
)

# -------------------------------------------------------------------------
# 5. EJECUCI√ìN (CON CORRECCI√ìN DE EMERGENCIA SI FALLA)
# -------------------------------------------------------------------------
println("\nüèÉ RUNNING GRID SEARCH FOR PCA...")

# Hacemos un chequeo de seguridad antes de llamar a la funci√≥n
if size(dataset_pca.x_train, 1) != size(dataset_pca.y_train_encoded, 1)
    println("‚ö†Ô∏è  ALERTA: Dimensiones no coinciden. Intentando transponer X...")
    # Si las filas no coinciden, transponemos X
    x_train_fixed = permutedims(dataset_pca.x_train)
    x_val_fixed   = permutedims(dataset_pca.x_val)
    x_test_fixed  = permutedims(dataset_pca.x_test)
    
    # Reconstruimos el dataset con X transpuesta
    dataset_pca = (
        x_train=x_train_fixed, y_train_encoded=dataset_pca.y_train_encoded, y_train_flux=dataset_pca.y_train_flux,
        x_val=x_val_fixed, y_val_encoded=dataset_pca.y_val_encoded,
        x_test=x_test_fixed, y_test_encoded=dataset_pca.y_test_encoded
    )
    println("   Nuevas dimensiones X Train: $(size(dataset_pca.x_train))")
end

approach_name="PCA"
# Ahora s√≠, llamamos a la funci√≥n
winner_tuple_pca = run_grid_search_v4(dataset_pca, approach_name, topologies_to_test, learning_rates_to_test)

metrics_object_pca = winner_tuple_pca[9]

println("\nüìä Generating Clinical Plots...")
generate_clinical_plots(metrics_object_pca, approach_name)

export_standardized_results_v2(winner_tuple_pca, dataset_pca, approach_name)
save_approach_results_v2(winner_tuple_pca, dataset_pca, approach_name)

println("\nüéâ SUCCESS: PCA Pipeline Finished!")
```

```julia
using MLJ
using MLJMultivariateStatsInterface

# -------------------------------------------------------------------------
# 1. CARGA Y PREPROCESAMIENTO (Z-SCORE)
# -------------------------------------------------------------------------
println("üöÄ INICIANDO APPROACH 3: ICA...")

# Cargar el modelo ICA expl√≠citamente para evitar conflictos
const MLJ_ICA = MLJ.@load ICA pkg=MultivariateStats verbosity=0

data_path = "heart_disease_uci.csv"
data, num_col, cat_col, target_col = load_and_clean_data(data_path)

# ICA necesita Z-Score (Media 0, Varianza 1)
approach_ica = prepare_data(
    data, num_col, cat_col, target_col,
    norm_method=:zscore 
)

# -------------------------------------------------------------------------
# 2. SEPARACI√ìN DE COLUMNAS (NUM√âRICAS vs CATEG√ìRICAS)
# -------------------------------------------------------------------------
# prepare_data devuelve las num√©ricas primero, luego las categ√≥ricas.
n_num = length(num_col) # 5 columnas num√©ricas

# Desempaquetar
x_train_raw = approach_ica.x_train
x_val_raw   = approach_ica.x_val
x_test_raw  = approach_ica.x_test

# Cortar Matrices
# Entrenamiento
x_num_train = x_train_raw[:, 1:n_num]      # Solo n√∫meros (para ICA)
x_cat_train = x_train_raw[:, n_num+1:end]  # Solo categor√≠as (se quedan igual)

# Validaci√≥n
x_num_val   = x_val_raw[:, 1:n_num]
x_cat_val   = x_val_raw[:, n_num+1:end]

# Test
x_num_test  = x_test_raw[:, 1:n_num]
x_cat_test  = x_test_raw[:, n_num+1:end]

println("   Separaci√≥n completada: $(n_num) features num√©ricas aisladas.")

# -------------------------------------------------------------------------
# 3. ENTRENAMIENTO ICA
# -------------------------------------------------------------------------
k_components = 5 # Intenta con 5 (m√°ximo de columnas num√©ricas)
println("   Fitting ICA (k=$k_components) on numerical data...")

# SOLUCI√ìN: Aumentamos maxiter y tol para que converja s√≠ o s√≠
ica_model = MLJ_ICA(outdim=k_components, maxiter=50000, tol=0.5)

# Entrenamos SOLO con los num√©ricos del Train Set
ica_machine = machine(ica_model, MLJ.table(x_num_train))

# Usamos try-catch por si acaso sigue fallando, para que no rompa todo
try
    MLJ.fit!(ica_machine, verbosity=0)
    println("   ‚úÖ ICA Converged successfully!")
catch e
    println("   ‚ö†Ô∏è ICA Warning: $e")
    println("   Continuing with best effort...")
end

# -------------------------------------------------------------------------
# 4. TRANSFORMACI√ìN Y RE-UNIFICACI√ìN
# -------------------------------------------------------------------------
# A. Transformar la parte num√©rica
x_num_train_ica = MLJ.matrix(MLJ.transform(ica_machine, MLJ.table(x_num_train)))
x_num_val_ica   = MLJ.matrix(MLJ.transform(ica_machine, MLJ.table(x_num_val)))
x_num_test_ica  = MLJ.matrix(MLJ.transform(ica_machine, MLJ.table(x_num_test)))

# B. Pegar de nuevo con la parte categ√≥rica (Concatenaci√≥n Horizontal)
# Convertimos todo a Matriz para que Flux no se queje
x_train_final_ica = hcat(x_num_train_ica, x_cat_train)
x_val_final_ica   = hcat(x_num_val_ica,   x_cat_val)
x_test_final_ica  = hcat(x_num_test_ica,  x_cat_test)

println("   ‚úÖ ICA Completado. Nueva forma X Train: $(size(x_train_final_ica))")
```

```julia
# -------------------------------------------------------------------------
# 5. PUENTE ANN (Adaptaci√≥n de Tipos)
# -------------------------------------------------------------------------
dataset_ica = (
    # ENTRENAMIENTO (Samples x Features) - Float32
    x_train         = Float32.(x_train_final_ica),
    
    # Targets Originales (BitMatrix)
    y_train_encoded = approach_ica.y_train_ohe,
    y_train_flux    = Float32.(approach_ica.y_train_ohe), # Backup
    
    # VALIDACI√ìN
    x_val           = Float32.(x_val_final_ica),
    y_val_encoded   = approach_ica.y_val_ohe,
    
    # TEST
    x_test          = Float32.(x_test_final_ica),
    y_test_encoded  = approach_ica.y_test_ohe
)

println("‚úÖ Puente ICA construido.")

# -------------------------------------------------------------------------
# 6. EJECUCI√ìN DEL EXPERIMENTO (GRID SEARCH)
# -------------------------------------------------------------------------
topologies_to_test = [
    [8], [16], [32], [64],          
    [16, 8], [32, 16], [64, 32], 
    [16, 16], [32, 32]      
]
learning_rates_to_test = [0.05, 0.01, 0.005, 0.001]
approach_name = "ICA"

println("\nüèÉ RUNNING GRID SEARCH FOR ICA...")

# 1. Buscar Ganador (Usando tu funci√≥n V3/V4 corregida)
winner_tuple_ica = run_grid_search_v3(dataset_ica, approach_name, topologies_to_test, learning_rates_to_test)

metrics_object_ica = winner_tuple_ica[9]

# 2. Gr√°ficos Cl√≠nicos
println("\nüìä Generating Clinical Plots...")
generate_clinical_plots(metrics_object_ica, approach_name)

# 3. Exportar Estandarizado (V2)
export_standardized_results_v2(winner_tuple_ica, dataset_ica, approach_name)

# 4. Guardar Modelo F√≠sico (V2)
save_approach_results_v2(winner_tuple_ica, dataset_ica, approach_name)

println("\nüéâ SUCCESS: ICA Pipeline Finished!")
```

```julia
using DataFrames, CSV, Random, MLJ, Flux, Statistics

# Asumimos que ya tienes cargadas tus funciones de 'unit6-modelcrossvalidation.jl'
# y las otras utilidades (load_and_clean_data, stratified_holdOut, etc.)

# -------------------------------------------------------------------------
# 1. PREPARACI√ìN DE DATOS (APPROACH 4)
# -------------------------------------------------------------------------
println("üöÄ STARTING APPROACH 4: CROSS-VALIDATION PIPELINE...")

data_path = "heart_disease_uci.csv"
data, num_col, cat_col, target_col = load_and_clean_data(data_path)

Random.seed!(1234)

# A. Split Inicial (85% Dev / 15% Test)
# Usamos Pval=0.0 porque el validation set para CV sale del 85% restante
# Ptest=0.15 para el test final.
# El resto ser√° Dev (Train + Val impl√≠cito)
(idx_dev, idx_trash, idx_test) = stratified_holdOut(data[!, target_col], 0.0, 0.15) 

# Re-combinamos train y val del split para formar "Dev" (85%)
# stratified_holdOut devuelve train, val, test. Aqu√≠ "train" ser√° nuestro Dev.
# Nota: Ajusta esto si tu funci√≥n stratified_holdOut devuelve indices distintos.
# Asumiendo que devuelve (train_indices, val_indices, test_indices)
# y queremos que train_indices sea el 85%.

dev_data  = data[idx_dev, :]
test_data = data[idx_test, :]

println("   Data Split: Dev=$(nrow(dev_data)) (85%), Test=$(nrow(test_data)) (15%)")

# B. Preparar Inputs para universalCrossValidation1
dev_num = select(dev_data, num_col)
dev_cat = select(dev_data, cat_col)
dev_targets = dev_data[!, target_col]

# C. Generar √çndices de CV (5 Folds)
k_folds = 5
# Usamos tu funci√≥n crossvalidation (o stratified_cv si la tienes)
# Si no tienes una funci√≥n 'crossvalidation', puedes usar esta r√°pida:
function simple_cv_indices(targets, k)
    n = length(targets)
    # Simple fold generation (puedes mejorar esto con estratificaci√≥n si tienes la funci√≥n)
    folds = zeros(Int, n)
    # Asignar aleatoriamente
    shuffled_idx = randperm(n)
    fold_size = floor(Int, n/k)
    for i in 1:k
        start_i = (i-1)*fold_size + 1
        end_i = i == k ? n : i*fold_size
        folds[shuffled_idx[start_i:end_i]] .= i
    end
    return folds
end

cv_indices = simple_cv_indices(dev_targets, k_folds)
println("   CV Indices Generated ($k_folds folds).")


# -------------------------------------------------------------------------
# 2. GRID SEARCH CON CROSS-VALIDATION
# -------------------------------------------------------------------------
println("\nüîé STARTING GRID SEARCH (CV)...")

# Definimos el espacio de b√∫squeda
topologies = [
    [8], [16], [32], [64],          
    [16, 8], [32, 16], [64, 32], 
    [16, 16], [32, 32]      
]
learning_rates = [0.05, 0.01, 0.005, 0.001]

results_cv = []

for topo in topologies
    for lr in learning_rates
        
        # Configurar hiperpar√°metros
        hyperparams = Dict(
            :topology => topo,
            :learningRate => lr,
            :maxEpochs => 1000,
            :validationRatio => 0.2 # Para el split interno de ANN
        )
        
        # Llamar a tu funci√≥n Universal
        # Retorna (mean_acc, std_acc)
        mean_acc, std_acc = universalCrossValidation1(
            :ANN, 
            hyperparams, 
            dev_num, 
            dev_cat, 
            dev_targets, 
            cv_indices
        )
        
        println("   Topo: $topo | LR: $lr -> Mean Acc: $(round(mean_acc*100, digits=2))% (¬±$(round(std_acc*100, digits=2)))")
        
        push!(results_cv, (topo, lr, mean_acc, std_acc))
    end
end

# -------------------------------------------------------------------------
# 3. SELECCI√ìN Y ENTRENAMIENTO FINAL
# -------------------------------------------------------------------------
# Ordenar por Mean Accuracy (√≠ndice 3)
sort!(results_cv, by = x -> x[3], rev = true)
winner = results_cv[1]
(best_topo, best_lr, best_acc, _) = winner

println("\nüèÜ WINNER CONFIGURATION: Topo $best_topo, LR $best_lr (Acc: $(round(best_acc*100, digits=2))%)")

println("\nüèãÔ∏è TRAINING FINAL MODEL ON FULL DEV SET...")

# 1. Preprocesar TODO el Dev Set (Fit & Apply)
params_proc = fit_preprocess(Matrix{Float64}(dev_num), dev_cat)
X_dev_proc = Float32.(apply_preprocess(Matrix{Float64}(dev_num), dev_cat, params_proc))

classes = sort(unique(dev_targets))
y_dev_ohe = oneHotEncoding(dev_targets, classes)

# 2. Entrenar ANN Final (Sin validaci√≥n externa, usamos split interno para early stopping)
# Usamos trainClassANN directamente
final_ann, _, _, _ = trainClassANN(
    best_topo,
    (X_dev_proc, y_dev_ohe);
    validationDataset = (Array{Float32}(undef,0,0), falses(0,0)), # Sin val set expl√≠cito (o usa un trocito peque√±o)
    learningRate = best_lr,
    maxEpochs = 1000,
    showText = false
)

# -------------------------------------------------------------------------
# 4. EVALUACI√ìN FINAL (TEST SET)
# -------------------------------------------------------------------------
println("\n‚öñÔ∏è  FINAL EVALUATION ON TEST SET...")

# 1. Preprocesar Test Set (Usando par√°metros aprendidos en Dev)
test_num = select(test_data, num_col)
test_cat = select(test_data, cat_col)
y_test   = test_data[!, target_col]

X_test_proc = Float32.(apply_preprocess(Matrix{Float64}(test_num), test_cat, params_proc))
y_test_ohe  = oneHotEncoding(y_test, classes)

# 2. Predecir
raw_preds = final_ann(X_test_proc')' # Transponer para Flux
y_pred_bool = classifyOutputs(raw_preds)

# 3. M√©tricas
metrics = confusionMatrix(y_pred_bool, y_test_ohe)
println("   üéØ FINAL TEST ACCURACY: $(round(metrics.accuracy * 100, digits=2))%")
println("   üéØ FINAL TEST F1:       $(round(metrics.f_score, digits=3))")

# 4. Guardar (Opcional)
# JLD2.save("Modelo_ANN_CV_Final.jld2", "ann_model", final_ann)
```

```julia
using DataFrames, CSV, Random, MLJ, Flux, Statistics, MultivariateStats

# Asumimos que 'ann_utils.jl' ya tiene cargada la funci√≥n 'universalCrossValidation_PCA'
# que me mostraste anteriormente.

# -------------------------------------------------------------------------
# 1. PREPARACI√ìN DE DATOS (APPROACH 5: PCA + CV)
# -------------------------------------------------------------------------
println("üöÄ STARTING APPROACH 5: PCA + CROSS-VALIDATION PIPELINE...")

data_path = "heart_disease_uci.csv"
data, num_col, cat_col, target_col = load_and_clean_data(data_path)

Random.seed!(1234)

# A. Split Inicial (85% Dev / 15% Test)
# El 'Dev Set' se usar√° para Cross-Validation y entrenamiento final
(idx_dev, _, idx_test) = stratified_holdOut(data[!, target_col], 0.0, 0.15) 

dev_data  = data[idx_dev, :]
test_data = data[idx_test, :]

println("   Data Split: Dev=$(nrow(dev_data)) (85%), Test=$(nrow(test_data)) (15%)")

# B. Preparar Inputs
dev_num = select(dev_data, num_col)
dev_cat = select(dev_data, cat_col)
dev_targets = dev_data[!, target_col]

# C. Generar √çndices de CV (5 Folds)
k_folds = 5
cv_indices = simple_cv_indices(dev_targets, k_folds) # Tu funci√≥n auxiliar
println("   CV Indices Generated ($k_folds folds).")


# -------------------------------------------------------------------------
# 2. GRID SEARCH (CON PCA SEGURO)
# -------------------------------------------------------------------------
println("\nüîé STARTING GRID SEARCH (PCA-CV)...")

# Espacio de b√∫squeda

topologies = [
    [8], [16], [32], [64],          
    [16, 8], [32, 16], [64, 32], 
    [16, 16], [32, 32]      
]
learning_rates = [0.05, 0.01, 0.005, 0.001]
pca_components = 15 # O el n√∫mero que decidas (ej: 15 explica ~95%)

results_cv = []

for topo in topologies
    for lr in learning_rates
        
        hyperparams = Dict(
            :topology => topo,
            :learningRate => lr,
            :maxEpochs => 1000,
            :validationRatio => 0.2, # Split interno de ANN para Early Stopping
            :pca_components => pca_components # Pasamos k para el PCA interno
        )
        
        # Llamamos a tu funci√≥n SEGURA que hace PCA dentro de cada fold
        mean_acc, std_acc = universalCrossValidation_PCA(
            :ANN, 
            hyperparams, 
            dev_num, 
            dev_cat, 
            dev_targets, 
            cv_indices
        )
        
        println("   Topo: $topo | LR: $lr | PCA: $pca_components -> Acc: $(round(mean_acc*100, digits=2))%")
        
        push!(results_cv, (topo, lr, mean_acc))
    end
end

# -------------------------------------------------------------------------
# 3. ENTRENAMIENTO FINAL (RE-TRAINING)
# -------------------------------------------------------------------------
# Ordenar por Accuracy
sort!(results_cv, by = x -> x[3], rev = true)
winner = results_cv[1]
(best_topo, best_lr, best_acc) = winner

println("\nüèÜ WINNER: Topo $best_topo, LR $best_lr (CV Acc: $(round(best_acc*100, digits=2))%)")
println("\nüèãÔ∏è PIPELINE FINAL (Fit Preprocess -> Fit PCA -> Train ANN)...")

# --- A. PREPROCESAMIENTO GLOBAL (Z-Score + OHE) ---
# Aprendemos los par√°metros usando TODO el Dev Set (85%)
params_proc = fit_preprocess_zscore(Matrix{Float64}(dev_num), dev_cat)

# Aplicamos a Dev y Test
X_dev_hyb  = Float32.(apply_preprocess_zscore(Matrix{Float64}(dev_num), dev_cat, params_proc))
test_num   = select(test_data, num_col)
test_cat   = select(test_data, cat_col)
X_test_hyb = Float32.(apply_preprocess_zscore(Matrix{Float64}(test_num), test_cat, params_proc))

# --- B. PCA GLOBAL ---
# Aprendemos la matriz de rotaci√≥n PCA usando TODO el Dev Set preprocesado
# Importante: Transponer para MultivariateStats (Features x Samples)
pca_m = MultivariateStats.fit(MultivariateStats.PCA, X_dev_hyb'; maxoutdim=pca_components)

# Proyectamos Dev y Test al espacio reducido
X_dev_final  = Float32.(MultivariateStats.transform(pca_m, X_dev_hyb')') # (Samples x 15)
X_test_final = Float32.(MultivariateStats.transform(pca_m, X_test_hyb')') # (Samples x 15)

println("   PCA Dimension Reduction: $(size(X_dev_hyb, 2)) -> $(size(X_dev_final, 2)) features")

# --- C. ENTRENAMIENTO ANN ---
classes = sort(unique(dev_targets))
y_dev_ohe = oneHotEncoding(dev_targets, classes)

# Entrenamos la red final con los datos reducidos
final_ann, train_h, _, _ = trainClassANN(
    best_topo,
    (X_dev_final, y_dev_ohe); # X ya tiene dimensiones correctas (Samples x Features)
    validationDataset = (Array{Float32}(undef,0,0), falses(0,0)), 
    learningRate = best_lr,
    maxEpochs = 1000,
    showText = false
)

# -------------------------------------------------------------------------
# 4. EMPAQUETADO Y EXPORTACI√ìN
# -------------------------------------------------------------------------
println("\nüì¶ PACKAGING RESULTS FOR EXPORT...")

# Preparamos los targets del test para las m√©tricas
y_test_targets = test_data[!, target_col]
y_test_ohe = oneHotEncoding(y_test_targets, classes)

# Construimos un 'dataset' artificial para reutilizar tus funciones de guardado
# Importante: X_test_final ya es (Samples x Features), tus funciones V2 esperan eso y lo transponen.
dataset_pca_cv = (
    x_test = X_test_final,       
    y_test_encoded = y_test_ohe,
    x_train = X_dev_final,       # Solo por referencia
    y_train_encoded = y_dev_ohe
)

# Creamos una tupla ganadora falsa solo para pasar los datos a la funci√≥n
# (Topo, Acc, F1, Sens, Spec, Prec, Time, MODELO)
# Calculamos m√©tricas r√°pidas para llenar los huecos
preds = final_ann(X_test_final')
y_p_bool = classifyOutputs(preds)
m = confusionMatrix(y_p_bool, y_test_ohe)
dummy_winner = (best_topo, m.accuracy*100, m.f_score, m.sensitivity, m.specificity, m.ppv, 0.0, final_ann)

# 1. Gr√°ficos
println("\nüìä Generating Clinical Plots...")
generate_clinical_plots(m, "PCA-CV")

# 2. Exportar
export_standardized_results_v2(dummy_winner, dataset_pca_cv, "PCA_CV")
save_approach_results_v2(dummy_winner, dataset_pca_cv, "PCA_CV")

println("\nüéâ Approach 5 (PCA+CV) Completed!")
```
